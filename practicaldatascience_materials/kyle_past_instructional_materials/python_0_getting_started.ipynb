# Getting Started

This introduction will guide you in terms of how to get started using the components of this course: the Spyder Integrated Development Environment (IDE) and provide some guidance on how to use Jupyter Notebooks.

In this course you'll be programming in Python. Python just takes text files with the .py extension, so you can create them anywhere, but Spyder, as we'll see, is a power tool for making programming easier. Jupyter Notebooks are a different development environment for Python that you'll use for each of the exercises that you complete through this course - while most of the content will be presented through Jupyter Notebooks, you'll do your actualy programming work in Spyder.

These tools all come with the [Anaconda](https://www.anaconda.com/download/) distribution of Python. While you can install that for free on your computer, we provide a place for work through a virtual machine(VM), which is just a computer that you can access remotely. We'll walk you through how to launch the VM and subsequently how to launch Spyder in the VM. Lastly, we'll discuss how to use Jupyter Notebooks.

# Jupyter Notebooks

All of the lessons in this course are presented through Jupyter Notebooks, as you'll see through descriptions later, Jupyter Notebooks combine text, code, and the output of functions all in once place.

While Jupyter notebooks can be run on multiple systems (it's installed on the VM, for examples), we've created a place for you to work with Jupyter Notebooks from one simple link. The link [https://vm-manage.oit.duke.edu/containers/mids](https://vm-manage.oit.duke.edu/containers/mids) brings you to a Jupyter Notebook environment you can use (you may need to enter you netid and password):

<img src='./img/0.11.png'>

You'll see a folder called "jupyter-r-python-notebooks." Click on this, then you'll see notebooks for each of the modules associated with the MIDS Online Review. Go to the Python folder, and this will present you with a list of notebooks for this course (instructions for how to proceed through the course are on Sakai). 

For the vast majority of the course, you'll be using Spyder - following the content in the Jupyter Notebooks, and trying them yourself in Spyder (this will help you practice and build your coding skills). Jupyter Notebooks also provide an interactive way of working with Python code as well. Before you edit a notebook, make a copy of it first so you don't edit the original content. From the directory with your files, check the box next to the notebook you want to copy, then click the "duplicate" button on the top. This will produce a copy. Click on that copy to open the notebook.

<img src='./img/0.12.png'>

There's a lot you can do with Jupyter Notebooks, so I'll mention a few features here. First off, notebooks are divided up into cells which can contain either text (using what's known as markdown language - it's a way of indicating how you want the text styled) and code (which can be executed right in the notebook). The example below shows a notebook with three cells: one with text, and two with code:

<img src='./img/jupyter_overview1.png'>

There are a number of features of the notebook. The buttons along the top control things like copying and pasting cells, moving cells around, and executing the code in the cells. These tools operate on selected cells. If your cursor is currently flashing in a cell, that cell is selected. Additionally, if you click on the side of a cell, it turns blue (as shown in the figure above) and you have selected it. You can hold down shift to select a range of cells. You can even make plots in the Jupyter notebooks.

<img src='./img/jupyter_overview2.png'>

If you wanted to return to the folder, you can click the Jupyter Notebook logo in the top left.

That should get you started with the core tools for the course. You're ready to get started!
